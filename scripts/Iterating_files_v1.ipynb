```python
#create function that does everything, from filtering which raw_data to import, to select corpus
#data and all the other functions in merging_data

def combine(sub, corpus_df):
    """
    :param sub: int, subject number
    :corpus_df: complete Potsdam Search Corpus dataframe
    
    :return:
    """
    
    raw_data = pd.read_csv("../separate_participant_data/asc\SFC"+str(sub)+"_2.dat", sep = "\t", 
                       header = None, names=['time', 'x', 'y', 'pupil'])
    
    corpus_sub = corpus_df[corpus_df.subject == sub] #filter corpus data to get only 
                                                     #from that participant
        
    raw_valid_times,intervals = get_valid_times(raw_data, corpus_sub)
    data_dva = linear_transf_pix_dva(raw_valid_times)
    data_trial_info, trial_durations = associate_trial_info(data_dva, intervals, corpus_sub)
    final = add_sacc_val_id(data_trial_info, corpus_sub)
    
    
     
    final.to_csv('../separate_participant_data/raw/raw_subj_'+str(sub)+'.dat',index=False,
                         columns=list(final.columns))

    return
    
```

```python
%%time
for idx,sub in enumerate(subs_list):
    print(idx)
    combine(sub,corpus_df)
```
